```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_wine
X,y  = load_wine(return_X_y = True)
```

```python
data = load_wine()
df  =pd.DataFrame(data.data,columns = data.feature_names)
df.head()
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn import tree
dt = tree.DecisionTreeClassifier()
dt.fit(X,y)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize = (24,14))
tree.plot_tree(dt,filled = True,fontsize = None)
```

```python
clf = tree.DecisionTreeClassifier(min_impurity_decrease=0.0,min_samples_split=40)
clf.fit(X,y)
plt.figure(figsize = (18,10))
tree.plot_tree(clf,filled = True , fontsize = 14)
```

```python
clf = tree.DecisionTreeClassifier(max_leaf_nodes=5)
clf.fit(X, y)
plt.figure(figsize=(20,10))
tree.plot_tree(clf, filled=True, fontsize=14)
```

```python
clf = tree.DecisionTreeClassifier(max_depth=3,min_samples_leaf=2)
clf.fit(X, y)
plt.figure(figsize=(20,10))
tree.plot_tree(clf, filled=True, fontsize=14)
```
